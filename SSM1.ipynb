```python
from midiutil import MIDIFile 

midi = MIDIFile(1) 
midi.addTrackName(0, 0, "Track") 
midi.addTempo(0, 0, 120) 
midi.addNote(0, 0, 60, 0, 4, 100) #longer note, duration = 4 

with open("test.mid", "wb") as f: midi.writeFile(f) 

#Remove old file 
import os
def safe_wav_path(folder, name="output.wav"):
    os.makedirs(folder, exist_ok=True)
    return os.path.abspath(os.path.join(folder, name))
    
if os.path.exists("output.wav"): 
    os.remove("output.wav")
```

```python
out_dir = r"C:\midi_output"

wav_path = safe_wav_path(out_dir)
midi_path = os.path.join(out_dir, "test.mid")


import subprocess 

sf2 = r"C:\Users\DELL\Downloads\GeneralUser_GS_v2.0.3--doc_r6\GeneralUser-GS\GeneralUser-GS.sf2" 
result = subprocess.run([
    r"C:\Users\DELL\Downloads\fluidsynth-v2.5.4-win10-x64-cpp11\fluidsynth-v2.5.4-win10-x64-cpp11\bin\fluidsynth.exe",
    "-ni",
    "-F", wav_path,
    "-T", "wav",
    "-r", "44100",
    sf2,
    "test.mid"
],
capture_output=True, text=True)

#delay 
import time 
time.sleep(0.5)

print("RETURN CODE:", result.returncode)
print("STDERR:", result.stderr)
print("STDOUT:", result.stdout)
print("EXPECTING WAV AT:", wav_path)
print("EXISTS?:", os.path.exists(wav_path))
```
